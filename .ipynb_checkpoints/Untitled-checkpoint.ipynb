```python
import pandas as pd

titanic = pd.read_csv("train.csv")
titanic
```

```python
# Let's remove Name and Ticket since there is no way to visualize this data
column_names = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']

# Drops missing values
titanic = titanic[column_names].dropna()
titanic
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.distplot(titanic["Fare"])
plt.show()
```

```python
sns.distplot(titanic["Age"])
plt.show()
```

```python
# Generating a kernel density plot for the "Age" column, with the area under the curve shaded
sns.kdeplot(titanic["Age"])

# Adds an x-axis label called "Age"
plt.xlabel("Age")
plt.show()
```

```python
# Sets the style sheet to coordinate grid hidden, and background color white
sns.set_style('white')

# Generates a kernel density plot with shaded area
sns.kdeplot(titanic['Age'], shade=True)

# Remove all spines
sns.despine(left=True, bottom=True)

# Make xlabel to "Age"
plt.xlabel('Age')
plt.show()
```

```python
# Condition on unique values of the "Survived" column.
g = sns.FacetGrid(titanic, col="Survived", size=6)
# For each subset of values, generate a kernel density plot of the "Age" columns.
g.map(sns.kdeplot, "Age", shade=True)
plt.show()
```

```python
# Use the FacetGrid instance to generate three plots on the "Pclass" row
g2 = sns.FacetGrid(titanic, col='Pclass', size=6)
g2.map(sns.kdeplot, "Age", shade=True)  # Drawing a kernel density plot of the "Age" column

# Remove all of the spines
sns.despine(left=True, right=True, top=True, bottom=True)

plt.show()
```

```python
# Use a FacetGrid instance to generate a grid of plots using the following conditions:
# 1. The Survived column across the columns in the grid.
# 2. The Pclass column across the rows in the grid.
# 3. The Sex column using different hues.
g3 = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", size=6)

g3.map(sns.kdeplot, "Age", shade=True)

# Adds a legend
g3.add_legend()   

# Remove all spines
sns.despine(top=True,bottom=True,left=True,right=True)
plt.show()
```
